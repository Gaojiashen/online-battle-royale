# Project Vision
## 这是什么游戏？
这是一款受《密教模拟器》启发的多人在线文字沙盒游戏。

玩家以「求道者」的身份进入一个充满隐秘知识、超自然力量与派系斗争的动态世界，通过与NPC、势力互动，逐步接触更高层次的秘密，并最终尝试完成属于自己的飞升。

游戏通过私聊进行，玩家向法官提交行动，法官依据规则书推动世界运行并裁定结果。

玩家既是世界的参与者，也是世界历史的创造者。
## 为什么好玩？
### 第一层：探索未知

玩家进入世界时几乎一无所知。

他们不知道：

哪些势力存在

哪些NPC掌握秘密

哪些事件即将发生

飞升的具体方式

他们需要主动探索。

这种乐趣来自：

「世界比我知道的更大。」

### 第二层：发现机会

随着探索深入，玩家开始接触各种机会。

例如：

某个帮派正在招募成员

某位学者掌握禁忌知识

某个遗迹即将开启

某个组织正在筹备仪式

玩家需要判断：

这是否值得投入？

乐趣来自：

「我发现了别人还不知道的东西。」

### 第三层：影响世界

玩家不是旁观者。

他们的行为会改变局势。

例如：

帮派战争的胜负

势力的兴衰

NPC的命运

某个区域的控制权

世界会因为玩家的行动发生变化。

乐趣来自：

「这件事是我推动发生的。」

### 第四层：竞争飞升

随着飞升进度推进。

玩家会逐渐意识到：

我不是唯一的求道者。

其他玩家：

可能与自己合作

可能争夺同一资源

可能阻止自己的仪式

可能提前飞升

于是竞争自然产生。

乐趣来自：

「我必须在他之前抵达真相。」
## 核心体验是什么？

在一个不断变化的神秘世界中，通过探索秘密、操纵局势、积累力量与知识，最终完成属于自己的飞升。

# Core Loop

探索世界

获取情报与秘密

发现飞升机会

采取行动争夺机会

获得知识、资源或影响力

推进飞升进程

暴露自身并引发新的风险与竞争
